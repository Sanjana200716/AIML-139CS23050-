```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Downloads/titanic.csv')
df=pd.DataFrame(df)
df.head()
```

```python
df.isnull().sum()
```

```python
mean=df['age'].mean()
df['age']=df['age'].fillna(mean)
df['age']
```

```python
df.isnull().sum()
```

```python
mode=df['age'].mode()
df['age']=df['age'].fillna(mode)
df['age']
```

```python
df.isnull().sum()
```

```python
median=df['age'].median()
df['age']=df['age'].fillna(median)
df['age']
```

```python
df.isnull().sum()
```

```python
df['embarked']=df['embarked'].fillna('S')
df['embarked']
```

# => read the titanic dataset  and perform the following operations:
    1}.Identify the outliers using scatter plot or box plot.
    2}.Replace the outliers using mean value and median value.
    3}.Replace the outliers using upper cap and lower cap.
    4}.Remove the data objects with outliers.

```python
import pandas as pd
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1+1.5*IQR
upper_bound=Q3-1.5*IQR
outliers=(df['age']<lower_bound) | (df['age']>upper_bound)
df_median=df.copy()
median_df=df['age'].median()
df_median.loc[outliers,'age']=median_df
df['age']
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('Downloads/titanic.csv')
df=pd.DataFrame(df)
fare_arr=df['fare']
lower_cap=np.percentile(fare_arr,1)
upper_cap=np.percentile(fare_arr,99)
for i in df['fare']:
    if i <lower_bound:
        df['fare']=df['fare'].replace(i,lower_cap)
    elif i> upper_bound:
        df['fare']=df['fare'].replace(i,upper_cap)
        print(df['fare'])
        
```
