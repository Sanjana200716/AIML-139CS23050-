```python
data=pd.read_csv("Desktop/Iris.csv")
df=pd.DataFrame(data)
df
df.isnull().sum()
```

```python
mean=df['SepalLengthCm'].mean()
df['SepalLengthCm']=df['SepalLengthCm'].fillna(mean)
df
```

```python
median=df['SepalWidthCm'].median()
df['SepalWidthCm']=df['SepalWidthCm'].fillna(median)
df
```

```python
median=df['PetalLengthCm'].median()
df['PetalLengthCm']=df['PetalLengthCm'].fillna(median)
df['PetalLengthCm']
```

```python
df.isnull().sum()
```

```python
df
```

```python
a=df.groupby('Species')['SepalLengthCm'].mean()
plt.scatter(a.index,a.values)
plt.xlabel("Species")
plt.ylabel("SepalLengthCm")
plt.title("Scatter plot")
```

```python
b=df.groupby('Species')['SepalWidthCm'].mean()
b.plot(kind='hist')
plt.xlabel("Species")
plt.ylabel("SepalWidthCm")
plt.title("Histogram")
```

```python
c=df.groupby('Species')['PetalLengthCm'].mean()
c.plot(kind='bar')
plt.xlabel("Species")
plt.ylabel("PetalLengthCm")
plt.title("Bar Graph")
```

```python
c=df.groupby('Species')['SepalWidthCm'].mean()
c.plot(kind='box')
plt.title("Boxplot")
```

```python
df
```

```python
d=df.groupby('Species')['SepalLengthCm'].mean()
d.plot(kind='pie',autopct='%1.2f%%')
```

```python
import seaborn as sns
c=df.groupby('Species')['PetalLengthCm'].mean()
sns.countplot(c)
```

```python
c=df.groupby('Species')['PetalLengthCm'].mean()
c.plot(kind='line')
plt.xlabel("Species")
plt.ylabel("PetalLengthCm")
plt.title("LineGraph")
```
